1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).


```python
import matplotlib.pyplot as plt
import random #importing because it is part of the standard library
import numpy as np #importing because it is included with matplotlib
from collections import Counter
TEXT = "At three o’clock precisely I was at Baker Street, but Holmes had not yet returned. The landlady informed me that he had left the house shortly after eight o’clock in the morning. I sat down beside the fire, however, with the intention of awaiting him, however long he might be. I was already deeply interested in his inquiry, for, though it was surrounded by none of the grim and strange features which were associated with the two crimes which I have already recorded, still, the nature of the case and the exalted station of his client gave it a character of its own. Indeed, apart from the nature of the investigation which my friend had on hand, there was something in his masterly grasp of a situation, and his keen, incisive reasoning, which made it a pleasure to me to study his system of work, and to follow the quick, subtle methods by which he disentangled the most inextricable mysteries. So accustomed was I to his invariable success that the very possibility of his failing had ceased to enter into my head.".lower()
letter_freq = dict(Counter(TEXT))

# make a bar plot with an appropriate title, correct xtick labels, and labeled axes
sorted_letter_freq = dict(sorted(letter_freq.items()))
plt.bar(sorted_letter_freq.keys(), sorted_letter_freq.values())
plt.xlabel('Letter')
plt.ylabel('Frequency')
plt.title('Letter Frequency in Sherlock Holmes')
plt.xticks(list(sorted_letter_freq.keys()))
plt.show()
```

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
# Generate a list of 5 random numbers for the y-axis
random_numbers_y = random.sample(range(100), 5)

# Generate a list of 5 random numbers for the x-axis
random_numbers_x = list(range(1, 6))

# Create the scatter plot
plt.scatter(random_numbers_x, random_numbers_y)
plt.xlabel('Index')
plt.ylabel('Random Y values')
plt.title('Scatter Plot of Random Numbers')
plt.show()

# Create a line plot using the same 5 random numbers as the scatter plot
plt.plot(random_numbers_x, random_numbers_y)
plt.xlabel('Index')
plt.ylabel('Random Y values')
plt.title('Line Plot of Random Numbers')
plt.show()

```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
# Create the scatter plot with different colors
plt.scatter(random_numbers_x, random_numbers_y, color='red', s=100)
plt.xlabel('Index')
plt.ylabel('Random Y values (mm)')
plt.title('Scatter Plot of Random Numbers')

# Create a second y-axis
ax2 = plt.gca().twinx()
ax2.set_ylabel('Random Y values (inch)')
ax2.set_ylim(plt.gca().get_ylim())
ax2.set_yticks(plt.gca().get_yticks())
ax2.set_yticklabels([f'{y/25.4:.2f}' for y in plt.gca().get_yticks()])  # Convert mm to inches

plt.grid(True)
plt.show()

# Create a line plot with different colors
plt.plot(random_numbers_x, random_numbers_y, color='red', linewidth=5)
plt.xlabel('Index', loc='left')
plt.ylabel('Random Y values', loc='top')
plt.gca().yaxis.set_label_position("right")
plt.gca().yaxis.tick_right()
plt.title('Line Plot of Random Numbers')
plt.grid(True)
plt.show()
```

3(b) I wanted to explore other ways to theme matplotlib plots.  So I am going to import an external library called mplcyberpunk as it is something that I have wanted to play with.

```python
import mplcyberpunk
plt.style.use("cyberpunk")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Create the scatter plot with the cyberpunk style
ax1.scatter(random_numbers_x, random_numbers_y)
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Y values (mm)')
ax1.set_title('Scatter Plot of Random Numbers')
ax1.grid(True)
mplcyberpunk.make_scatter_glow(ax1)

# Create the line plot with the cyberpunk style
ax2.plot(random_numbers_x, random_numbers_y)
ax2.set_xlabel('Index')
ax2.set_ylabel('Random Y values (mm)')
ax2.set_title('Line Plot of Random Numbers')
ax2.grid(True)
mplcyberpunk.add_glow_effects(ax2)

plt.tight_layout()
plt.show()
```

4. Given the data sets below, plot or scatter both on the same set of axes.  You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

```python
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# First plot
ax1.plot(sizes, insertion_sort_times, label='Insertion Sort', marker='o')
ax1.plot(sizes, merge_sort_times, label='Merge Sort', marker='x')
ax1.set_xlabel('List Size')
ax1.set_ylabel('Time (seconds)')
ax1.set_title('Sorting Algorithm Performance')
ax1.legend()
ax1.grid(True)
mplcyberpunk.add_glow_effects(ax1)

# Second plot with logarithmic x-axis
ax2.plot(sizes, insertion_sort_times, label='Insertion Sort', marker='o')
ax2.plot(sizes, merge_sort_times, label='Merge Sort', marker='x')
ax2.set_xlabel('List Size')
ax2.set_ylabel('Time (seconds)')
ax2.set_xscale('log')
ax2.set_title('Sorting Algorithm Performance (Logarithmic Scale)')
ax2.legend()
ax2.grid(True, which="both", ls="--")
mplcyberpunk.add_glow_effects(ax2)

plt.tight_layout()
plt.show()
```
